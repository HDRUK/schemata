# Notebook used for rough work when putting together schema for data utility

```python
import json
import requests
import re
```

Starting with dataset schema as a template

```python
dataset_schema = requests.get('https://raw.githubusercontent.com/HDRUK/schemata/master/schema/dataset/latest/dataset.schema.json').json()
dataset_schema.keys()
```

Adding basic top level info

```python
utility_schema = dataset_schema.copy()
utility_schema['$id'] = 'https://raw.githubusercontent.com/HDRUK/schemata/master/schema/data_utility/0.0.1/data_utility.schema.json'
utility_schema['title'] = 'HDR UK Data Utility Framework Evaluation Schema'
utility_schema['description'] = 'Schema for an evaluation against the HDR UK Data Utility Framework'
utility_schema['version'] = '0.0.1'
```

# Definitions - just one:

```python
# Pick an enum as a template:
enum_ = dataset_schema['definitions']['periodicity'].copy()
enum_['enum'] = ['platinum', 'gold', 'silver', 'bronze']

utility_schema['definitions'] = {'evaluation': enum_}
```

# Properties
https://www.hdruk.ac.uk/wp-content/uploads/2021/08/Data-Utility-Framework_withlink_updatedDC100821.pdf


```python
# First define the categories:
categories = ['dataDocumentation','technicalQuality',
    'coverage','accessAndProvision','valueAndInterest']

utility_schema['properties'] = {c:{} for c in categories}

def camel_to_title(string):
    return re.sub('([A-Z])', r' \1', string).title()

for p in utility_schema['properties'].keys():
    utility_schema['properties'][p]['$id'] = '#/properties/{}'.format(p)
    utility_schema['properties'][p]['title'] = camel_to_title(p)
    utility_schema['properties'][p]['description'] = ''
    utility_schema['properties'][p]['$ref'] = '#/definitions/{}'.format(p)
```
