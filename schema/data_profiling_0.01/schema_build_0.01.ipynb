# Properties


```python
# First define the categories:
categories = ['pid','technicalQuality',
    'coverage','accessAndProvision','valueAndInterest']

data_profile_schema['properties'] = {c:{} for c in categories}

def camel_to_title(string):
    return re.sub('([A-Z])', r' \1', string).title()

for p in data_profile_schema['properties'].keys():
    data_profile_schema['properties'][p]['$id'] = '#/properties/{}'.format(p)
    data_profile_schema['properties'][p]['title'] = camel_to_title(p)
    data_profile_schema['properties'][p]['description'] = ''
    data_profile_schema['properties'][p]['$ref'] = '#/definitions/{}'.format(p)
```
